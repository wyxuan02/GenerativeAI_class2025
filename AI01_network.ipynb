#第一週作業 02/18
**# 畫一個函數圖形**


｜實作說明｜

1. 建立 y = 2sin(x) 函式和對數函式作比較
2. 找出兩條函式的交點
3. 美化圖表：修改線條顏色、線條粗度、標示說明

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 定義 x 值範圍
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)

# 定義函數
y1 = 2 * np.sin(x)
y2 = np.log(np.abs(x) + 1) # 使用對數函數，並避免對數輸入為0

# 找出交點
intersection_points = []
for i in range(len(x) -1):
    if (y1[i] - y2[i]) * (y1[i+1] - y2[i+1]) < 0:
        intersection_points.append((x[i] + x[i+1])/2)

# 畫圖
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='y1 = 2sin(x)', color='blue', linewidth=3)
plt.plot(x, y2, label='y2 = log(|x|+1)', color='red', linewidth=3)

# 標記交點
for point in intersection_points:
    plt.plot(point, 2*np.sin(point), marker='o', markersize=8, color='green')


# 添加標題和標籤
plt.title('Comparison of $y1=2sin(x)$ and $y2=log(|x|+1)$', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# 添加圖例
plt.legend(fontsize=12)

# 設定網格線
plt.grid(True, linestyle='--', alpha=0.7)

# 顯示圖表
plt.show()
```
