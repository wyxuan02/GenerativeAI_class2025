# 期末專案主題：健康餐食推薦聊天機器人

**工具：**

連結 OpenAI 並使用 RAG 作為回應，最後以 Gradio Web App 呈現。

使用 Groq 金鑰，模型為 "llama3-70b-8192"。

**設計概念：**



*   個人化健康管理

    * 根據使用者輸入的年齡、性別、體重、活動量與目標（減脂、增肌、控糖等）客製化推薦三餐。

    * 提供 TDEE（總每日能量消耗）與營養素比例建議。

    * 參考來源：https://www.peeta.tw/all/tdee-and-bmr/

*   即時營養知識回覆（透過 RAG）

    * 支援查詢如「高蛋白低碳的晚餐有哪些？」、「什麼食物適合糖尿病患者？」。

    * 來源為衛生福利部國民健康署，透過檢索增強生成的方式回覆具體建議與科學依據。

    * 其餘菜單由 ChatGPT 生成建議。

    * 參考來源：https://www.hpa.gov.tw/Pages/List.aspx?nodeid=36

*   互動式菜單構建器

    * 使用者可輸入「我有雞胸肉、番茄和糙米」，系統回覆適合的健康食譜與營養分析。

    * 根據熱量限制（如「給我一份500卡的低脂午餐」）自動產生餐點。

*   過敏與飲食偏好篩選

    * 如「我對堅果過敏，請排除相關食材」或「我吃全素，請推薦適合的食譜」。
